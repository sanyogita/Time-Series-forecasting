# `Naive method`: assummes last entry is carried forward


```python
test_naive=test.copy()
test_naive['naive_forecast']=train.cnt.iloc[train_len-1]
```

```python
plt.plot(train.cnt, label='train')
plt.plot(test_naive.cnt, label='test')
plt.plot(test_naive.naive_forecast, label='naive_forcast')
plt.title("Naive forecast")
plt.legend(loc='best')
plt.show()
```

`Simple average method`: average of cnt over entire train period

```python
test_sa=test.copy()
test_sa['sa_forecast']=train.cnt.mean()
```

```python
plt.plot(train.cnt, label='train')
plt.plot(test_sa.cnt, label='test')
plt.plot(test_sa.sa_forecast, label='sa_forecast')
plt.title("Simple average")
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
rmse=np.sqrt(mean_absolute_error(test_naive.linear_interpolate,test_naive.naive_forecast)).round(2)
rmse
```

```python
mape=np.mean((np.abs(test_naive.linear_interpolate-test_naive.naive_forecast)/test_naive.linear_interpolate)*100).round(2)
rmse,mape
```

```python
result=pd.DataFrame({"Method":["Naive method"], "RMSE":[rmse], "MAPE":[mape]})
result
```

```python
rmse=np.sqrt(mean_absolute_error(test_sa.linear_interpolate,test_sa.sa_forecast)).round(2)
rmse
```

```python
mape=np.mean((np.abs(test_sa.linear_interpolate-test_sa.sa_forecast)/test_sa.linear_interpolate)*100).round(2)
mape
```

```python
tempResult=pd.DataFrame({'Method':"Simple Average","RMSE":[rmse],"MAPE":[mape]})
tempResult
```

```python
results=pd.concat([result,tempResult])
```

```python
from scipy.stats import boxcox
```

```python
data_boxcox=pd.Series(boxcox(data.linear_interpolate, lmbda=0),index=data.index)
```

```python
plt.plot(data_boxcox)
plt.title("After boxcox transformation")
```

```python
data_box_cox_diff=data_boxcox-data_boxcox.shift()
data_box_cox_diff.dropna(inplace=True)
```

```python
plt.plot(data_box_cox_diff)
```

```python
test_result=adfuller(data_box_cox_diff)
print("ADF test statistics:",test_result[0])
print("Critical value of 0.05 is ",test_result[4]['5%'])
print("P value of statistics: ",test_result[1])
```

```python
test_result=kpss(data_box_cox_diff)
print("kpss test statistics:",test_result[0])
print("Critical value of 0.05 is ",test_result[3]['5%'])
print("P value of statistics: ",test_result[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_acf(data_box_cox_diff, ax=plt.gca(), lags=30)
```

```python
plot_pacf(data_box_cox_diff, ax=plt.gca(), lags=30)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# train test split
```

```python
train_boxcox=data_boxcox[:train_len]
test_boxcox=data_boxcox[train_len:]
train_boxcox_diff=data_box_cox_diff[:train_len-1]
test_boxcox_diff=data_box_cox_diff[train_len-1:]
```

```python
model=ARIMA(train_boxcox_diff, order=(1,0,0)).fit()
```

```python
y=data.copy()
```

```python
y['boxcox_diff']=model.predict(data_box_cox_diff.index.min(),data_box_cox_diff.index.max())
y['boxcox']=y.boxcox_diff.cumsum()
y['boxcox']=y['boxcox'].add(data_boxcox[0])
y['forecast']=np.exp(y.boxcox)
```

```python
plt.plot(train.linear_interpolate, label='train')
plt.plot(y.forecast[test.index.min():],label='predicted')
plt.plot(test.linear_interpolate, label='test')
plt.legend(loc='best')
plt.title("Autoregression model")
```

```python
rmse=np.sqrt(mean_absolute_error(y.linear_interpolate[test.index.min():],y.forecast[test.index.min():]))
rmse
```

```python
mape=np.mean((np.abs(y.linear_interpolate[test.index.min():]-y.forecast[test.index.min():])/y.linear_interpolate[test.index.min():])*100)
mape
```

```python
tempResults=pd.DataFrame({"Method":["Auto Regression"], "RMSE":[rmse],'MAPE':[mape]})
results=pd.concat([results,tempResults])
results
```
