```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("UNRATE.csv", header=None)
```

```python
data.columns=["date",'rate']
data
```

```python
data['date']=pd.to_datetime(data.date, format="%d-%m-%Y")
data.index=data.date
```

```python
plt.figure(figsize=(12,4))
data.rate.plot()
plt.title("Rates raw")
```

```python
# BoxCOx transformation
from scipy.stats import boxcox

data_boxcox = pd.Series(boxcox(data.rate, lmbda=0),index=data.index)
```

```python
plt.figure(figsize=(12,4))
data_boxcox.plot()
plt.title("Rates boxcoxed")
```

```python
# Applying differenciation
data_boxcox_diff=pd.Series(data_boxcox-data_boxcox.shift(),data.index)
```

```python
data_boxcox_diff.dropna(inplace=True)
data_boxcox_diff.shape
```

```python
plt.figure(figsize=(12,4))
data_boxcox_diff.plot()
plt.title("Rates boxcoxed")
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss
```

```python
adf_test = adfuller(data_boxcox_diff)

print('ADF Statistic: %f' % adf_test[0])
print('Critical Values @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

```python
kpss_test = kpss(data_boxcox_diff)

print('KPSS Statistic: %f' % kpss_test[0])
print('Critical Values @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %f' % kpss_test[1])
```

```python
length=612
train=data_boxcox_diff[:length]
test=data_boxcox_diff[length:]
test
```

```python
data_train=data[:length]
data_test=data[length:]
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model=ARIMA(train, order=(12,0,0)) 
model_fit=model.fit()
print(model_fit.params)
```

```python
yhat_ar=data_boxcox_diff.copy()
yhat_ar['forecast_diff_boxcox']=model_fit.predict(test.index.min(),test.index.max())
```

```python
yhat_ar['forecast_boxcox']=yhat_ar['forecast_diff_boxcox'].cumsum()
yhat_ar['forecast_boxcox']=yhat_ar['forecast_boxcox'].add(data_boxcox[0])
yhat_ar['forecast']=np.exp(yhat_ar['forecast_boxcox'])
yhat_ar['forecast'].shape
```

```python
plt.figure(figsize=(12,4))
plt.plot(data_train.rate,label='train')
plt.plot(data_test.rate,label="test")
plt.plot(yhat_ar['forecast'], label="forecast")
plt.title("Rates boxcoxed")
plt.legend(loc="best")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(data_test.rate, yhat_ar['forecast'][data_test.index.min():])).round(2)
mape = np.round(np.mean(np.abs(data_test.rate,yhat_ar['forecast'][data_test.index.min():])/data_test.rate)*100,2)

tempResults = pd.DataFrame({'Method':['Autoregressive (AR) method'], 'RMSE': [rmse],'MAPE': [mape] })
tempResults
```
